# В этом файлике будет реализован метод наименьших квадратов для аппроксимации данных линейной зависимостью лабораторной работы 123. 

Необходимые импорты:

```python
from typing import Sequence

import matplotlib.pyplot as plt
import numpy as np
```

__Вбейте значения температуры в точке и шаг значений по x:__

```python
T_list = [33.1, 36.3, 39.4, 41.6, 43.4, 45.4, 47.7, 49.8, 52.3, 54.9]
# [31.9, 32.32, 32.86, 33.88, 34.95, 41.86, 42.14, 42.68, 43.60, 44.51, 55, 55.36, 55.79, 56.66, 57.48]
x_step = 2

abscissa = np.array(T_list)
```

__Вбейте значения сопротивления в точке и шаг значений по y:__

```python
R_list = [3.9620, 3.9978, 4.0431, 4.0730, 4.1065, 4.1310, 4.1597, 4.1806, 4.2211, 4.2606]
# [0.02755, 0.02697, 0.022732, 0.02699, 0.02671, 0.02743, 0.02768, 0.02749, 0.02735, 0.02732, 0.02824, 0.02808, 0.0279, 0.02766, 0.02726]
y_step = 0.01

ordinates_experimental = np.array(R_list)
```

__Запустите следующий код, чтобы увидеть график завсисмости R от T__

```python
# расчет данных для МНК

xy_mean = np.mean(abscissa * ordinates_experimental)

x_mean = np.mean(abscissa)
y_mean = np.mean(ordinates_experimental)
prod_of_xmean_ymean = x_mean*y_mean

mean_of_powered_abscissa = np.mean(abscissa**2)

powered_x_mean = x_mean**2

a = (xy_mean - x_mean*y_mean)/(mean_of_powered_abscissa - powered_x_mean)
b = y_mean - a*x_mean

# функция для графика МНК

def visualize_lsm(
    abscissa: np.ndarray,
    ordinates_experiment: np.ndarray,
    ordinates_theory: np.ndarray,
    x_step: float,
    y_step: float
) -> None:
    _, axis = plt.subplots(figsize=(14, 10))
    axis.set_title("MNK", fontsize=17, fontweight="bold", c="dimgray")

    axis.scatter(
        abscissa,
        ordinates_experiment,
        color="red",
        alpha=1,
        label = "экспериментальные значения"
    );

    axis.plot(abscissa, ordinates_theory, alpha=0.5, label="аппроксимация")

    
    plt.grid(True)
    plt.xticks(np.arange(np.min(abscissa) - x_step, np.max(abscissa) + x_step, x_step))
    plt.yticks(np.arange(np.min(ordinates_experiment) - y_step, np.max(ordinates_experiment) + y_step, y_step))

    axis.legend()
```

```python
# визуализация МНК
print ("зависисмость R от T")
visualize_lsm(
    abscissa=abscissa,
    ordinates_experiment=ordinates_experimental,
    ordinates_theory=abscissa*a + b,
    x_step = x_step,
    y_step = y_step
)
```

__Запустите этот следующий код, чтобы узнать дополнительные данные об эксперименте__

```python
print (f"Функция принимаает следующий вид: R = {np.round(a, 4)} * T + {np.round(b, 4)}\n")
print (f"a = {np.round(a, 4)}, b = {np.round(b, 4)}\n")
```

```python
Tmean_list = np.array([37.37, 38.22, 38.49, 39.62, 41.25, 44.17, 44.23, 44.89, 45.98, 46.81, 52.32, 52.68, 53.12, 53.12, 55.19])
x_step_ =  1

abscissa_ = np.array(Tmean_list)
```

__Вбейте значения коэвициента теплопроводности воздуха в точке и шаг значений по y:__

```python
K_exp = np.array([0.03041, 0.02760, 0.02875, 0.02792, 0.02558, 0.02688, 0.02843, 0.02750, 0.02624, 0.02693, 0.02790, 0.02727, 0.02700, 0.02700, 0.02593])
y_step_ = 0.0001
ordinates_experimental_ = np.array(K_exp)
```

__Запустите следующий код, чтобы увидеть график завсисмости коэфициента теплопроводности среды от средней температуры__

```python
# расчет данных для МНК

xy_mean_ = np.mean(abscissa_ * ordinates_experimental_)

x_mean_ = np.mean(abscissa_)
y_mean_ = np.mean(ordinates_experimental_)
prod_of_xmean_ymean_ = x_mean_*y_mean_

mean_of_powered_abscissa_ = np.mean(abscissa_**2)

powered_x_mean_ = x_mean_**2

a_ = (xy_mean_ - x_mean_*y_mean_)/(mean_of_powered_abscissa_ - powered_x_mean_)
b_ = y_mean_ - a_*x_mean_

# ФУНКЦИЯ ВИЗУАЛИЗАЦИИ

def visualize_lsm_(
    abscissa: np.ndarray,
    ordinates_experiment: np.ndarray,
    ordinates_theory: np.ndarray,
    ordinates_th: np.ndarray,
    x_step: float,
    y_step: float
) -> None:
    _, axis = plt.subplots(figsize=(14, 10))
    axis.set_title("K(T)", fontsize=17, fontweight="bold", c="dimgray")

    axis.scatter(
        abscissa,
        ordinates_experiment,
        color="red",
        alpha=1,
        label = "экспериментальные значения"
    );

    axis.scatter(
        abscissa,
        ordinates_th,
        color="green",
        alpha=1,
        label = "табличные значения"
    );

    axis.plot(abscissa, ordinates_theory, alpha=1, label="аппроксимация")

    
    axis.grid(True)
    plt.xticks(np.arange(np.min(abscissa_), np.max(abscissa_), x_step_))
    plt.yticks(np.arange(np.min(ordinates_experimental_) - y_step_*5, np.max(ordinates_experimental_) + y_step_ * 7, y_step))

    axis.legend()
```

```python
ordinates_th_list = [0.02698, 0.02705, 0.02707, 0.02716, 0.02729, 0.02751, 0.02752, 0.02757, 0.02766, 0.02772, 0.02815, 0.02818, 0.02821, 0.02821, 0.02837]
ordinates_th = np.array(ordinates_th_list)

print ("зависимость К от Тср")
visualize_lsm_(
    abscissa=abscissa_,
    ordinates_experiment=ordinates_experimental_,
    ordinates_theory=abscissa_*a_ + b_,
    ordinates_th = ordinates_th,
    x_step = x_step_,
    y_step = y_step_
)
```
